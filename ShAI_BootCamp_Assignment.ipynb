# 3- CLEANING THE DATA:

If I have a lage dataset, I can choose to just get rid of the rows that have missing values, zeros, or any unwanted information. On the other hand, if the dataset is small, I might decide to fill in the missing values with the average or middle value (mean or median). In this case, since the dataset is large, I've opted to remove the rows with unwanted information.

```python
#No of missing values in each column
df.isna().sum()
```

```python
#Drop columns that are empty
df.drop(['Notes','Status'], axis=1, inplace=True)
print(df.columns,df.shape)
```

```python
#Dropping rows where any of the specified columns contains zero(its meaningless zeros)
columns_to_check = ['BasePay', 'TotalPay','TotalPayBenefits']

#Use boolean indexing to filter rows where any of the specified columns contain zero
df.drop(df[df[columns_to_check].eq(0).any(axis=1)].index, inplace=True)
df.shape
```

```python
#Drop rows where any of the specified columns contain missing values (NaN)
df.dropna(subset=['Benefits','BasePay', 'TotalPay','TotalPayBenefits'], inplace=True)
df.shape
```

```python
#Rechick to see no of missing values in each column
df.isna().sum()
```

```python
#Drop rows where any of the specified columns contain negative values (NaN)

negcolumns = ['BasePay', 'OvertimePay', 'OtherPay']

# Use boolean indexing to filter rows where any of the specified columns contain a negative value
df.drop(df[df[negcolumns].lt(0).any(axis=1)].index, inplace=True)
df.shape
```

```python
#Create the clean dataset
df.to_csv('cleaned dataset.csv', index=False)
```
